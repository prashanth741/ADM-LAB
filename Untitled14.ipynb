```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('customer_segmentation.csv')

features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
x = df[features]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(x_scaled)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['Customer Segmentation'] = df['Customer Segmentation']
print(pca_df.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('customer_segmentation.csv')

features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
x = df[features]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(x_scaled)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['Customer Segmentation'] = df['Customer Segmentation']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Customer Segmentation', palette='Set2', s=70)
plt.title('PCA of Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Segment')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_stock_data.csv")

features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_10']
x = df[features]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Company'] = df['Company']

print(pca_df.head())

plt.figure(figsize=(8, 6))
for company in pca_df['Company'].unique():
    subset = pca_df[pca_df['Company'] == company]
    plt.scatter(subset['PC1'], subset['PC2'], label=company)
plt.title('PCA of Stock Market Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df = pd.read_csv("sample_stock_data.csv")
sector_map = {
    'AlphaCorp': 'Technology',
    'BetaTech': 'Healthcare'
}
df['Sector'] = df['Company'].map(sector_map)


features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA_5', 'MA_10']
x = df[features]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)

pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pca_df['Sector'] = df['Sector']

plt.figure(figsize=(8, 6))
for sector in pca_df['Sector'].unique():
    subset = pca_df[pca_df['Sector'] == sector]
    plt.scatter(subset['PC1'], subset['PC2'], label=sector)
plt.title('PCA of Stock Market Data by Sector')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
