```python
import pandas as pd
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(url)
data.head()
```

```python
print(data.isnull().sum())
data['Age'].fillna(data['Age'].median(), inplace=True)                                                                                                                                                            data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True) # Verify that missing values are handled
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_normalize = ['Age', 'Fare']
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
data[columns_to_normalize].head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_normalize = ['Age', 'Fare']
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
data[columns_to_normalize].head()
```

```python
if 'sepal_length' in data.columns:
    data['sepal_length'].fillna(data['sepal_length'].mean(), inplace=True)
else:
    print("Column 'sepal_length' not found in the DataFrame.")
```

```python
print("\nData Summary Statistics:")
print(data.describe())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from google.colab import files
data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print("Missing values before imputation:")
print(data.isnull().sum())
print("\nMissing values after imputation:")
print(data.isnull().sum())
scaler_minmax = MinMaxScaler()
columns_to_normalize = ['Age', 'Fare']
data[columns_to_normalize] = scaler_minmax.fit_transform(data[columns_to_normalize])
scaler_standard = StandardScaler()
data[columns_to_normalize] = scaler_standard.fit_transform(data[columns_to_normalize])
data.head()
```
